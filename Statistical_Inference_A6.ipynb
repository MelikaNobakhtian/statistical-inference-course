#Q1

Many college courses conclude by giving students the opportunity to evaluate the course and the instructor anonymously. However, the use of these student evaluations as an indicator of course quality and teaching effectiveness is often criticized because these measures may reflect the influence of non-teaching related characteristics, such as the physical appearance of the instructor. Researchers  at the  University  of Texas,  Austin collected  data  on  teaching  evaluation  score (higher score means better) and standardized beauty score (a score of 0 means average, a negative score means below average, and a positive score means above average) for a sample of 463 professors. The  scatterplot  below  shows  the  relationship  between  these  variables,  and  also provided is a regression output for predicting teaching evaluation score from beauty score.

- **a.**  Given that the average standardized beauty score is −0.0883 and the average teaching evaluation score is 3.9983, calculate the slope. Alternatively, the slope may be computed using just the information provided in the model summary table.

- **b.**  Do these data provide convincing evidence that the slope of the relationship between teaching evaluation and beauty is positive? Explain your reasoning.

Screenshot (1446).png

## Answer:

### Part (a): Calculate the Slope

We know that the regression equation is:
$ Y = \beta_0 + \beta_1 X $

Given:
- Average standardized beauty score ($\bar{X}$) = -0.0883
- Average teaching evaluation score ($\bar{Y}$) = 3.9983
- Intercept ($\beta_0$) = 4.010

We need to find the slope ($\beta_1$).

Since the regression line passes through the point ($\bar{X}$, $\bar{Y}$), we can substitute these values into the regression equation to solve for $\beta_1$:

$ \bar{Y} = \beta_0 + \beta_1 \bar{X} $

Substitute the given values:

$ 3.9983 = 4.010 + \beta_1 (-0.0883) $

Solve for $\beta_1$:

$ 3.9983 - 4.010 = \beta_1 (-0.0883) $
$ -0.0117 = \beta_1 (-0.0883) $
$ \beta_1 = \frac{-0.0117}{-0.0883} $
$ \beta_1 \approx 0.1325 $

So, the slope ($\beta_1$) is approximately 0.1325.

### Part (b): Evidence of Relationship

To determine whether there is convincing evidence that the slope of the relationship between teaching evaluation and beauty is positive, we need to examine the p-value associated with the slope coefficient in the regression summary table.

Given:
- T Value = 4.13
- p-value = 0.0000

The hypotheses for testing the slope are:
- **Null Hypothesis ($H_0$)**: The slope $\beta_1 = 0$. This means there is no relationship between beauty and teaching evaluation scores.
- **Alternative Hypothesis ($H_A$)**: The slope $\beta_1 \neq 0$. This means there is a relationship between beauty and teaching evaluation scores.

Since the p-value is 0.0000, which is significantly less than the common significance level ($\alpha = 0.05$), we reject the null hypothesis. The small p-value provides convincing evidence that the slope of the relationship between teaching evaluation and beauty is positive. This means that as the beauty score increases, the teaching evaluation score also tends to increase.



# Q2

The Coast Starlight Amtrak train runs from Seattle to Los Angeles. The scatterplot below displays the distance between each stop (in miles) and the amount of time it takes to travel from one stop to
another (in minutes).

The mean travel time from one stop to the next on the Coast Starlight is 129 mins, with a standard deviation of 113 minutes. The mean distance traveled from one stop to the next is 108 miles with a standard deviation of 99 miles. The correlation between travel time and distance is 0.636.
- **a.**   Write the equation of the regression line for predicting travel time.
- **b.**   Interpret the slope and the intercept in this context.
-**c.**   Calculate 𝑅2 of the regression line for predicting travel time from distance traveled for the Coast Starlight, and interpret 𝑅2 in the context of the application.
-**d.**   The distance between Santa Barbara and Los Angeles is 103 miles. Use the model to estimate the time it takes for the Starlight to travel between these two cities.
- **e.**   It actually takes the Coast Starlight about 168 mins to travel from Santa Barbara to Los Angeles. Calculate the residual and explain the meaning of this residual value.
- **f.**    Suppose Amtrak is considering adding a stop to the Coast Starlight 500 miles away from Los Angeles. Would it be appropriate to use this linear model to predict the travel time
from Los Angeles to this point?

Screenshot (1448).png

## Answer:



### Part (a): Equation of the Regression Line

To write the equation of the regression line, we use the formula:
$ Y = \beta_0 + \beta_1 X $

The slope ($\beta_1$) can be calculated using the formula:
$ \beta_1 = Cor(x,y) \left( \frac{s_y}{s_x} \right) $

Where:
- $Cor(x,y)$ is the correlation coefficient between travel time and distance.
- $s_y$ is the standard deviation of the travel time.
- $s_x$ is the standard deviation of the distance.

Given:
- $Cor(x,y) = 0.636$
- $s_y = 113$ minutes
- $s_x = 99$ miles

So:
$ \beta_1 = 0.636 \left( \frac{113}{99} \right) = 0.726 $

To find the intercept $(\beta_0$), we use the means of the travel time ($\bar{y}$) and distance ($\bar{x}$):
$ \beta_0 = \bar{y} - \beta_1 \bar{x} $

Given:
- $\bar{y} = 129$ minutes
- $\bar{x} = 108$ miles

So:
$ \beta_0 = 129 - (0.726 \times 108) = 129 - 78.408 = 50.592 $

Thus, the regression equation is:
$ \hat{y} = 50.592 + 0.726x $

### Part (b): Interpretation of the Slope and Intercept

- **Slope (0.726)**: For each additional mile traveled, the travel time is expected to increase by approximately 0.726 minutes.
- **Intercept (50.592)**: When the distance traveled is zero miles, the estimated travel time is approximately 50.592 minutes. This can be interpreted as the base time needed for stops and delays unrelated to distance.

### Part (c): Calculation of $R^2$

$R^2$ measures the proportion of the variance in the dependent variable (travel time) that is predictable from the independent variable (distance).

Given:
- $Cor(x,y) = R = 0.636$

So:
$ R^2 = (0.636)^2 = 0.404 $

**Interpretation**: Approximately 40.4% of the variance in travel time can be explained by the distance traveled.

### Part (d): Estimate Travel Time for 103 Miles

Using the regression equation:
$ \hat{y} = 50.592 + 0.726 \times 103 $

$ \hat{y} = 50.592 + 74.778 $

$ \hat{y} = 125.37 $

The estimated travel time for the 103 miles between Santa Barbara and Los Angeles is approximately 125.37 minutes.

### Part (e): Calculation of Residual

Residual is the difference between the actual travel time and the predicted travel time.

Given:
- Actual travel time = 168 minutes
- Predicted travel time = 125.37 minutes

Residual:
$ \text{Residual} = 168 - 125.37 = 42.63 $

**Explanation**: The residual of 42.63 minutes indicates that the actual travel time is 42.63 minutes longer than the time predicted by the model for the trip from Santa Barbara to Los Angeles.

### Part (f): Appropriateness of Using the Model for 500 Miles

Using the model to predict travel time for a distance significantly larger than the distances used to create the model (such as 500 miles) is often inappropriate due to the potential for extrapolation errors. The model was built based on data with a mean distance of 108 miles and a standard deviation of 99 miles, so predicting for a distance much larger than this range (more than 3 standard deviations away) can lead to inaccurate and unreliable predictions. Therefore, it would not be appropriate to use this linear model to predict the travel time for a distance of 500 miles.

## Answer:

**The equation of regression line:**

 ${y} = \beta_0 + \beta_1 x$

Where:
- ${y}$ is the predicted number of wins.
- $x$ is the goals scored per match.
- $\beta_0$ is the y-intercept.
- $\beta_1$ is the slope of the line.


### Step 1: Determine the Slope $(\beta_1)$

The slope $(\beta_1)$ can be calculated using two points from the regression line. Let's pick two points on the regression line for our calculation. Based on the plot:

- Point 1: (0.25, 0)
- Point 2: (0.75, 7)

The slope is calculated as:
$ \beta_1 = \frac{\Delta y}{\Delta x} = \frac{y_2 - y_1}{x_2 - x_1} $

Substitute the points into the formula:
$ \beta_1 = \frac{7 - 0}{0.75 - 0.25} = \frac{7}{0.5} = 14$

### Step 2: Determine the Intercept ($\beta_0$)

The intercept ($\beta_0$) is where the line crosses the y-axis (when $(x = 0)$). We can use one of the points on the line and the slope to find the intercept.

Using Point 1 (0.25, 0):

${y} = \beta_0 + \beta_1 x $

$ 0 = \beta_0 + 14 \times 0.25 $

$ 0 = \beta_0 + 3.5 $

$ \beta_0 = -3.5 $

### Step 3: Write the Linear Equation

Now that we have $\beta_1 = 14$ and $\beta_0 = -3.5$, the linear equation is:

$ {y} = -3.5 + 14x $

# Q9

The dataset prostate  is from a study of 97 men with prostate cancer who were due to receive a radical prostatectomy. Predict the lweight using the age.
- **a.**   Plot the data and comment on the relationship.
- **b.**   Fit a curve using kernel methods, plotting the fit on top of the data. What is the effect of
the outlier?
- **c.**Compute the smoothing spline fit with the default amount of smoothing. What type of
curve has been fit to the data?
- **d.** Fit a loess curve with a 95% confidence band. Do you think a linear fit is plausible for this
data?
- **e.**  Display all three previous fits on top of the same display and compare.
- **f.** Introduce lpsa as a second predictor and show the bivariate fit to the data using smoothing
splines.
- **g.**   Plot the residuals and interpret them.

## Answer:

### Load Dataset

```r
prostate <- read.table(
  'https://web.stanford.edu/~hastie/ElemStatLearn/datasets/prostate.data',
  header = TRUE)
```

```r
head(prostate)
```

### a)

From the scatter plot, we observe that there is no clear linear relationship between age and log weight. But we can observe a weak linear relationship that when the age increases, we will have also increase in log of weight:

```r
# Plot the data
ggplot(prostate, aes(x = age, y = lweight)) +
  geom_point(alpha = 0.5) +
  labs(title = "Relationship between Age and Log Weight",
       x = "Age",
       y = "Log Weight (lweight)")
```

### b)

The kernel smoothing fit is influenced by the outlier at the lower and higher end of the age spectrum. The deviation around these outliers indicates their impact on the fit:

```r
# Compute the kernel smoothed values
smoothed_values <- with(prostate, ksmooth(age, lweight, "normal", bandwidth = 5))

# Create a data frame for the smoothed values
smoothed_df <- data.frame(age = smoothed_values$x, smoothed_lweight = smoothed_values$y)

# Plot the data with kernel fit using ggplot2
ggplot(prostate, aes(x = age, y = lweight)) +
  geom_point(alpha = 0.5) +
  geom_line(data = smoothed_df, aes(x = age, y = smoothed_lweight), color = "green") +
  labs(title = "Kernel Smoothing Fit: Age vs. Log Weight",
       x = "Age",
       y = "Log Weight (lweight)")
```

### c)


The smoothing spline fit results in a simple line (linear) that adjusts to the overall trend of the data, balancing between capturing the general pattern and avoiding overfitting.


```r
# Fit the smoothing spline with default parameters
smoothing_spline_fit <- smooth.spline(age, lweight)

# Predict values using the fitted smoothing spline
predicted_spline_values <- predict(smoothing_spline_fit, age)$y

# Create scatter plot with smoothing spline fit
ggplot(prostate, aes(x=age, y=lweight)) +
  geom_point() +
  geom_line(aes(y=predicted_spline_values), color="red") +
  labs(title="Smoothing Spline Fit of lweight vs. Age",
       x="Age",
       y="lweight")
```

### d)

If the LOESS curve is roughly a straight line, a linear fit might be plausible. If the LOESS curve shows significant curvature, then a linear fit might not be appropriate. Here we have the second scenario, so it is not plausible:

```r
# Create scatter plot with LOESS fit and 95% confidence band
ggplot(prostate, aes(x=age, y=lweight)) +
  geom_point() +
  geom_smooth(method="loess", span=0.75, level=0.95, color="blue", fill="blue", alpha=0.2) +
  labs(title="LOESS Fit of lweight vs. Age with 95% Confidence Band",
       x="Age",
       y="lweight")
```

### e)

In the previous parts, we analyze different features of each fitted line. If we want to comapre them, it seems that the LOESS curve is the best one. This curve is robust to outliers (in contrast with ksmooth curve) and also it is not too general like spline fit:

```r
# Create scatter plot with LOESS fit, linear fit, and smoothing spline fit
ggplot(prostate, aes(x=age, y=lweight)) +
  geom_point() +
  geom_line(data = smoothed_df, aes(x = age, y = smoothed_lweight), color = "green") +
  geom_line(aes(y=predicted_spline_values), color="red") +
  geom_smooth(method="loess", span=0.75, level=0.95, color="blue", fill="blue", alpha=0.2)
  labs(title="Comparison of Kernel, Smoothing Spline and LOESS  Fits",
       x="Age",
       y="lweight") +
  theme_minimal() +
  theme(legend.position="top") +
  scale_color_manual(name="Fits", values=c("blue"="LOESS", "red"="Smoothing Spline", "green"="Kernel"))
```

### f)

```r
# Fit a smoothing spline with age and lpsa
library(mgcv)

# Fit model
spline_bivariate_fit <- gam(lweight ~ s(age, lpsa), data = prostate)

# Create a grid for age and lpsa
age_grid <- seq(min(prostate$age), max(prostate$age), length.out = 100)
lpsa_grid <- seq(min(prostate$lpsa), max(prostate$lpsa), length.out = 100)
grid <- expand.grid(age = age_grid, lpsa = lpsa_grid)
grid$lweight <- predict(spline_bivariate_fit, newdata = grid)

# Plot the bivariate fit
ggplot(grid, aes(x = age, y = lpsa, z = lweight)) +
  geom_tile(aes(fill = lweight)) +
  labs(title = "Bivariate Fit: Age and lpsa vs. Log Weight",
       x = "Age",
       y = "Log PSA (lpsa)") +
  scale_fill_gradientn(colors = terrain.colors(10)) +
  theme_minimal()
```

### g)

The residual plot shows that there are no clear patterns and most of the residuals have small values suggesting that the model has captured most of the
underlying structure in the data. However, there are some points with large residuals which may indicate outliers or areas where the model does not fit as well.

```r
# Compute residuals
prostate$residuals <- prostate$lweight - fitted(spline_bivariate_fit)

# Plot residuals
ggplot(prostate, aes(x = age, y = residuals)) +
  geom_point(alpha = 0.5) +
  geom_hline(yintercept = 0, linetype = "dashed") +
  labs(title = "Residuals of the Bivariate Fit: Age vs. Residuals",
       x = "Age",
       y = "Residuals")
```
