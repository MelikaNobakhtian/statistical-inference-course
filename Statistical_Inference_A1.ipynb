Test scores for a college statistics class held during the day are:

99,56,78,55.5,32,90,80,81,56,59,45,77,84.5,84,70,72,68,32,79,90

Test scores for a college statistics class held during the evening are:

98,78,68,83,81,89,88,76,65,45,98,90,80,84.5,85,79,78,98,90,79,81,25.5

```r
# Test scores for the day class
day_scores <- c(99, 56, 78, 55.5, 32, 90, 80, 81, 56, 59, 45, 77, 84.5, 84, 70, 72, 68, 32, 79, 90)

# Test scores for the night class
night_scores <- c(98, 78, 68, 83, 81, 89, 88, 76, 65, 45, 98, 90, 80, 84.5, 85, 79, 78, 98, 90, 79, 81, 25.5)
```

**(a):** Find the smallest and largest values, the median, and the first and third quartile for the day class and night class.

```r
# Smallest and largest values, median, and quartiles for the day class
day_min <- min(day_scores)
day_max <- max(day_scores)
day_median <- median(day_scores)
day_q1 <- quantile(day_scores, 0.25)
day_q3 <- quantile(day_scores, 0.75)

print(paste("Min for day class: ", day_min))
print(paste("Max for day class: ", day_max))
print(paste("Median for day class: ", day_median))
print(paste("First quartile for day class: ", day_q1))
print(paste("Third quartile for day class: ", day_q3))
```

```r
# Smallest and largest values, median, and quartiles for the night class
night_min <- min(night_scores)
night_max <- max(night_scores)
night_median <- median(night_scores)
night_q1 <- quantile(night_scores, 0.25)
night_q3 <- quantile(night_scores, 0.75)

print(paste("Min for night class: ", night_min))
print(paste("Max for night class: ", night_max))
print(paste("Median for night class: ", night_median))
print(paste("First quartile for night class: ", night_q1))
print(paste("Third quartile for night class: ", night_q3))
```

**(b):** Are there any outliers in any group? What are the exact values? Show the calculation
of detecting outliers.

To detect outliers, we use IQR method. At first, we should find the value for IQR: IQR is the difference of first quartile and third quartile. Now we can find lower bound and upper bound:

$\text{Lower Bound} =Q_1 - 1.5 \times \text{IQR}$

$\text{Upper Bound}=Q_3 + 1.5 \times \text{IQR}$

Every entry beyond these limits will be determined as outlier.

```r
# Define outliers using IQR method
day_iqr <- IQR(day_scores)
night_iqr <- IQR(night_scores)

day_outliers <- day_scores[(day_scores < (day_q1 - 1.5 * day_iqr)) | (day_scores > (day_q3 + 1.5 * day_iqr))]
night_outliers <- night_scores[(night_scores < (night_q1 - 1.5 * night_iqr)) | (night_scores > (night_q3 + 1.5 * night_iqr))]
```

For day classes, we don't have any outlier:

```r
day_outliers
```

For night classes, we have two outliers: 45, 25.5.

```r
night_outliers
```
