2. **Event B :** The sum of two dice equals 6.
   - Possible outcomes: (1, 5), (2, 4), (3, 3), (4, 2), (5, 1)

   - Total outcomes: 36
   - Probability of $ C :  P(C) = \frac{11}{36} $
   

### Conditional Probabilities:
a. **P(A|C):**
   - At first, we should find $P(A \cap C)$
   - $A \cap C$: (2, 2)
   - $ P(A \cap C) = \frac{1}{36} $
   - $ P(A|C) = \frac{P(A \cap C)}{P(C)} = \frac{\frac{1}{36}}{\frac{11}{36}} = \frac{1}{11} $
   
b. **P(B|C):**
   - At first, we should find $P(B \cap C)$
   - $B \cap C$: (2, 4), (4, 2).
   - $P(B \cap C) = \frac{2}{36}$
   - $P(B|C) = \frac{P(B \cap C)}{P(C)} = \frac{\frac{2}{36}}{\frac{11}{36}} = \frac{2}{11}$

### Independence:
A  and B are independent if $P(A|B) = P(A)$ and $P(B|A) = P(B)$.

a. **Independence of A  and  C :**

b. **Independence of B and  C :**
   - Since $P(B|C) = \frac{2}{11} $ and $ P(B) = \frac{5}{36} $, B and C are not independent.

# Q5

You write a software program over and over, and each time there is probability ùëù that it works correctly, independent of previous attempts. What is the mean and variance of ùëã, the number of tries until the program works correctly?

## Answer:

The number of tries until the program works correctly has a geometric distribution, which is models the number of trials needed to achieve the first success in a sequence of independent Bernoulli trials, where each trial has a constant probability of success $ p $.

For a geometric distribution, $\mu$ and $\sigma^2$ can be calculated as follows:

1. **$ \mu $**: The mean of a geometric distribution is given by the formula:
$\mu = \frac{1}{p} $

2. **$ \sigma^2 $**: The variance of a geometric distribution is given by the formula:
$\sigma^2 = \frac{1-p}{p^2} $

In this question, we will learn more about Monte Carlo simulation and understand how it can help us to model and analyze complex systems, where analytical or closed-form solutions are difficult. We are going to estimate the value of ùúã using Monte Carlo simulation. The idea is to randomly generate points within a square and determine how many fall in a circle inscribed by the square. By comparing the points inside the circle to the total number of points you can approximate ùúã. Repeat your simulation with different number of points and analyze their results.

## Answer:

```r
# Function to estimate pi using Monte Carlo simulation
estimate_pi <- function(num_points) {
  # Generate random points within the square [-1, 1] x [-1, 1]
  points <- matrix(runif(2 * num_points, min = -1, max = 1), ncol = 2)

  # Calculate the distance of each point from the origin
  distances <- sqrt(points[, 1]^2 + points[, 2]^2)

  # Count the number of points falling inside the inscribed circle (distance <= 1)
  inside_circle <- sum(distances <= 1)

  # Estimate pi using the ratio of points inside the circle to total points
  pi_estimate <- 4 * inside_circle / num_points

  return(pi_estimate)
}

# Number of points for simulation
num_points_list <- c(100, 1000, 10000, 100000, 1000000)

# Perform Monte Carlo simulation for different number of points
pi_estimates <- sapply(num_points_list, estimate_pi)

# Print the estimates of pi
for (i in seq_along(num_points_list)) {
  cat("Number of points:", num_points_list[i], "Estimated value of pi:", pi_estimates[i], "\n")
}

# Plot the estimates of pi against the number of points
plot(num_points_list, pi_estimates, type = "o", log = "x",
     xlab = "Number of Points", ylab = "Estimated Value of Pi",
     main = "Monte Carlo Simulation for Estimating Pi")
grid()
```

**Observations:**
* According to the reuslts, it is obvious that in most cases, if we increase the number of points that is our sample size, we will have a more accurate estimation of ùúã.
* With 1000000 sample size, we reached the value of 3.14 for ùúã, which is the value we usually use for calculations.
